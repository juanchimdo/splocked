```python
# web scrape
from splocked.utils import *
url = 'https://www.imdb.com/title/tt0111161/reviews?ref_=tt_ov_rt'
# turn into a df
df = get_reviews(url)
```

```python
# load the model
from tensorflow.keras.models import load_model
model = load_model('../../model_baseline')
```

```python
# load the word_to_dic
import json
with open('../../word_to_id.json') as json_file: 
    word_dict = json.load(json_file)
word_dict
```

```python
df.head()
```

```python
# Apply preprocessing 
from splocked.predict import preprocess_review
import numpy as np
preprocessed_reviews = [preprocess_review(review, word_dict) for review in df['comment']]
```

```python
# apply the prediction to all the reviews 
df['spoiler_proba'] = [model.predict(review)[0][0]*100 for review in preprocessed_reviews]
df.head(10)
```
